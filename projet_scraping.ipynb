```python
import requests
from bs4 import BeautifulSoup

from HtmlTags import HtmlTags
class Blog:
    def __init__(self,url):
        r = requests.get(url).text
        soup = BeautifulSoup(r, 'html.parser')
        self.title=soup.find('h2',class_='entry-summary__title').text
        self.date=soup.find('time').text
        self.author=soup.find('a',class_='author url fn').text
        self.auth_link=soup.find('a',class_='author url fn')['href']
        self.link=soup.find('a',class_='read-more')['href']
        self.text=soup.find('a',class_='read-more').parent.text
        tags = [child.name for child in soup.findChildren()]
        self.tags=HtmlTags(tags).getTags()

    def getInformation(self):
        return {
            'title':self.title,
            'date':self.date,
            'author':self.author,
            'author_link':self.auth_link,
            'link':self.link,
            'text':self.text,
            'tags':self.tags
        }

```

```python
import requests
from bs4 import BeautifulSoup

from Blog import Blog
from HtmlTags import HtmlTags


class Home:
    def __init__(self,url='https://m.signalvnoise.com/search/'):
        r = requests.get(url).text
        soup = BeautifulSoup(r, 'html.parser')
        ul = soup.ul
        tags = [child.name for child in soup.findChildren()]
        self.tags=HtmlTags(tags).getTags()
        L = ul.find_all('a')
        links = [[l.text, l['href']] for l in L]
        self.blogs=[Blog(link[1]).getInformation() for link in links]

    #getting articles from the search page
    def getInformation(self):
        return {
            'tags':self.tags,
            'blogs':self.blogs
        }

```

```python
class HtmlTags:
    def __init__(self,tags):
        self.tags=tags
    def getTags(self):
        Tags=[]
        for i in range(0,len(self.tags)):
            nbr=1
            for j in range(i+1,len(self.tags)):
                if((self.tags[j]  != '0') and (self.tags[i] == self.tags[j])):
                    nbr+=1;self.tags[j]='0'
            if(self.tags[i] != '0'):
                Tags.append({self.tags[i]:nbr})
        return Tags
```

```python
import requests
from bs4 import BeautifulSoup
import json

from Home import Home

page = Home().getInformation()
with open("BlogScrapping.json",'w') as file:
    file.write(json.dumps(page,sort_keys=True,indent=2))
    print("BlogScrapping.json is created in 'venv' folder....")
```
